# **STEP 2 — Load Summarization Model**

```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer
import torch

# Load the tokenizer and model for summarization
tokenizer = AutoTokenizer.from_pretrained("facebook/bart-large-cnn")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/bart-large-cnn")

def custom_summarizer(text, max_length, min_length, do_sample):
    inputs = tokenizer(text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(
        inputs["input_ids"],
        num_beams=4,
        max_length=max_length,
        min_length=min_length,
        do_sample=do_sample,
        early_stopping=True
    )
    # Decode the generated ids back to text
    decoded_summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    # Mimic the pipeline output format for consistency with existing calls
    return [{'generated_text': decoded_summary}]

# Assign the custom function to 'summarizer'
summarizer = custom_summarizer

print("Model loaded successfully")
```

# **STEP 3 — Prepare Dataset**

```python
texts = [

"""Artificial Intelligence is transforming industries worldwide. It is widely used in healthcare, finance, and education. Machine learning models analyze large datasets to identify patterns. However, ethical concerns and job displacement are major issues. Governments are regulating AI development.""",

"""The Indian monsoon plays a vital role in agriculture. Farmers depend on rainfall for crops. Climate change has caused irregular rainfall patterns, leading to floods and droughts. This affects food production and livelihoods.""",

"""Natural Language Processing allows computers to understand human language. It is used in chatbots, translation, and sentiment analysis. Challenges include ambiguity and context understanding.""",

"""A young boy from a small village dreamed of becoming a scientist. Despite poverty, he worked hard and achieved success, inspiring others.""",

"""During a meeting, the manager discussed deadlines and assigned tasks. Team members shared updates and proposed solutions. The meeting ended with clear responsibilities.""",

"""Technology is rapidly evolving with advancements in cloud computing and cybersecurity. Organizations adopt digital transformation but face cyber risks."""
]

print("Dataset prepared")
```

# **STEP 4 — Generate Summaries**

```python
for i, text in enumerate(texts):
    summary = summarizer(text, max_length=60, min_length=20, do_sample=False)

    print(f"\nText {i+1}")
    print("Original:", text)
    print("Summary:", summary[0]['generated_text'])
```

# **TEP 5 — Control Summary Length**

```python
sample = texts[0]

short = summarizer(sample, max_length=30, min_length=10, do_sample=False)
medium = summarizer(sample, max_length=60, min_length=20, do_sample=False)
long = summarizer(sample, max_length=100, min_length=40, do_sample=False)

print("Short:", short[0]['generated_text'])
print("\nMedium:", medium[0]['generated_text'])
print("\nLong:", long[0]['generated_text'])
```
