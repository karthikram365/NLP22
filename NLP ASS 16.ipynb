## Step 2 — Sentence Processing

```python
sentences = [
"The cat sits on the mat.",
"John bought a new car.",
"The teacher explained the lesson clearly.",
"Students are studying for the exam.",
"Artificial Intelligence is transforming industries.",
"Did you finish the assignment?",
"The book was written by the author.",
"The quick brown fox jumps over the lazy dog.",
"She enjoys reading novels.",
"The project was completed successfully."
]
for sentence in sentences:
    doc = nlp(sentence)
    print("Sentence:", sentence)
    for token in doc:
        print(token.text, token.pos_, token.dep_)
    print("\n")
```

## Step 3 — Extract Dependency Relations

```python
import pandas as pd

sentence = "The quick brown fox jumps over the lazy dog."
doc = nlp(sentence)

data = []

for token in doc:
    data.append([
        token.text,
        token.lemma_,
        token.pos_,
        token.dep_,
        token.head.text
    ])

df = pd.DataFrame(data, columns=["Token","Lemma","POS","Dependency","Head Word"])
print(df)
```

## Step 5 — Dependency Tree Visualization

```python
from spacy import displacy

sentence = "The quick brown fox jumps over the lazy dog."
doc = nlp(sentence)

displacy.render(doc, style="dep", jupyter=True)
```

## Step 6 — Comparative Analysis

Simple sentences produced short dependency trees with fewer branches.

Complex sentences contained multiple clauses, resulting in deeper trees.

Adjective modifiers like quick and lazy added extra nodes to the tree.

Prepositional phrases such as over the dog expanded the structure.

Passive sentences changed dependencies where the object becomes the subject.

For example, in passive voice nsubjpass appears instead of nsubj.

Sentences with conjunctions created additional dependency relations.

Dependency trees help clearly identify grammatical structure.
