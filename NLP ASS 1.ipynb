```python
# Initialize the Porter Stemmer
ps = PorterStemmer()
# Apply stemming to each token in the 'tokens' list
stemmed_tokens = [ps.stem(word) for word in tokens]
# Print the original tokens for comparison
print(f"Tokens: {tokens}")
# Print the resulting stemmed tokens
print(f"Stemmed tokens: {stemmed_tokens}")
```
