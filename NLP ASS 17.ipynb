# **STEP 2 — Load Translation Model**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
tokenizer_fr = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-fr")
model_fr = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-fr")

def translate_fr(text):
    inputs = tokenizer_fr(text, return_tensors="pt")
    outputs = model_fr.generate(**inputs)
    return tokenizer_fr.decode(outputs[0], skip_special_tokens=True)
```

```python
tokenizer_de = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-de")
model_de = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-de")

def translate_de(text):
    inputs = tokenizer_de(text, return_tensors="pt")
    outputs = model_de.generate(**inputs)
    return tokenizer_de.decode(outputs[0], skip_special_tokens=True)
```

```python
tokenizer_hi = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-hi")
model_hi = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-en-hi")

def translate_hi(text):
    inputs = tokenizer_hi(text, return_tensors="pt")
    outputs = model_hi.generate(**inputs)
    return tokenizer_hi.decode(outputs[0], skip_special_tokens=True)
```

# **STEP 3 — Prepare Input Sentences**

```python
sentences = [
"I am learning Natural Language Processing.",
"The weather is very pleasant today.",
"Artificial intelligence is transforming the world.",
"Do you like machine learning?",
"The meeting will start at 10 AM.",
"Students are preparing for their exams.",
"Technology is evolving rapidly.",
"I enjoy reading books about science.",
"Can you help me solve this problem?",
"The professor explained the concept clearly.",
"Social media influences modern communication.",
"Python is widely used in data science.",
"The research paper discusses neural networks.",
"We should always keep learning new skills.",
"How does machine translation work?"
]
```

# **STEP 4 — Perform Translation**

```python
for sentence in sentences:
    fr = translate_fr(sentence)
    print("English:", sentence)
    print("French :", fr)
    print()
```

# **STEP 5 — Translate into Multiple Languages**

```python
for sentence in sentences:
    fr = translate_fr(sentence)
    de = translate_de(sentence)
    hi = translate_hi(sentence)

    print("English:", sentence)
    print("French :", fr)
    print("German :", de)
    print("Hindi  :", hi)
    print("------------------------------------------------")
```

# **STEP 6 — Translation Quality Analysis **

Machine translation generally produces grammatically correct sentences when translating simple statements. However, complex sentences sometimes lead to slight meaning shifts. Technical terms are usually preserved correctly because pretrained models have seen similar vocabulary during training. French and German translations tend to maintain sentence structure closer to English, while Hindi translations often change word order due to different grammatical rules. Some idioms or cultural expressions may not translate perfectly. Questions and conversational sentences are handled well, though long sentences occasionally reduce fluency. Overall, pretrained models provide accurate translations for most general sentences but may struggle with domain-specific or ambiguous language.

# **STEP 7 — Testing Complex Sentences**



```python
complex_sentences = [
"Although artificial intelligence is powerful, it still requires human supervision.",
"If students practice coding regularly, they will improve their problem solving skills.",
"Machine learning models can analyze large datasets that humans cannot process quickly."
]

for s in complex_sentences:
    print("English:", s)
    print("French :", translate_fr(s))
    print("German :", translate_de(s))
    print()
```

1. What is a Sequence-to-Sequence model?

A Sequence-to-Sequence (Seq2Seq) model is a neural network architecture that converts one sequence into another sequence, such as translating a sentence from one language to another.

2. What is machine translation?

Machine translation is the automatic conversion of text from one language to another using computational algorithms and AI models.

3. What is encoder–decoder architecture?

Encoder–decoder architecture consists of two parts:

Encoder: Converts input text into a numerical representation.

Decoder: Generates the translated output from that representation.

4. What is a Hugging Face pipeline?

A Hugging Face pipeline is a high-level interface that allows users to easily perform NLP tasks such as translation, summarization, and sentiment analysis using pretrained models.

5. Why are pretrained models used in NLP?

Pretrained models save time and resources because they are already trained on large datasets and can perform tasks effectively without requiring additional training.

6. Limitations of automatic translation

Difficulty with idioms and cultural phrases

Errors in complex sentences

Incorrect translation of rare technical terms

Context misunderstanding in long paragraphs

7. Real-life applications

Google Translate and multilingual communication tools

International business and document translation

The translation pipeline allows quick conversion of English text into multiple languages such as French, German, and Hindi. Although the models provide accurate translations for most sentences, some limitations remain when dealing with complex grammar, idioms, and context-sensitive expressions.
